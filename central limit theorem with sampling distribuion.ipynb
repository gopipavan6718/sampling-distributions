```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def binomial():
    print("Population parameters pf binomial distribution are:population_size=1000,probability_of_success=0.3")
    population_size = 10000
    probability_of_success = 0.5 

    population = np.random.binomial(10, probability_of_success, population_size)
    ax=sns.displot(population,kde=True)
    population_mean = np.mean(population)
    population_variance = np.var(population)
    print("Mean and variance of population are:",population_mean,population_variance)

    print(" Number of samples to take from the population are 1000")
    num_samples = 1000

    print("Size of each sample is 50")
    sample_size = 50

    sample_means=[]

    for x in range(1,1000):
        sample=np.random.choice(population,sample_size)
        sample_means.append(np.mean(sample))


    sns.displot(sample_means, kde=True)
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.title('Sample Means (Central Limit Theorem using Binomial Distribution)')
    plt.grid(True)
    plt.show()
    
    sampling_mean = np.mean(sample_means)
    sampling_variance = np.var(sample_means)
    sampling_std = np.std(sample_means)
    sampling_stderr = sampling_std / np.sqrt(sample_size)
    print(f'Sampling Mean: {sampling_mean}')
    print(f'Sampling Variance: {sampling_variance}')
    print(f'Sampling Standard Deviation: {sampling_std}')
    print(f'Sampling Standard Error: {sampling_stderr}')
```

```python
def exponential():
    print("Population parameters of exponential distribution are population_size of 10000,population_mean of 5")
    population_size = 10000
    population_mean = 5

    population = np.random.exponential(population_mean, population_size)
    ax=sns.displot(population,kde=True)
    population_mean = np.mean(population)
    population_variance = np.var(population)
    print("Mean and variance of population are:",population_mean,population_variance)
    print("Number of samples to take from the population are 1000")
    num_samples = 1000

    print("Size of each sample is 50")
    sample_size = 50
    sample_means=[]
    for x in range(num_samples):
        sample=np.random.choice(population,sample_size)
        sample_means.append(np.mean(sample))



    sns.displot(sample_means, kde=True)

    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.title('Sample Means (Central Limit Theorem using Exponential Distribution)')
    plt.grid(True)
    plt.show()
    sampling_mean = np.mean(sample_means)
    sampling_variance = np.var(sample_means)
    sampling_std = np.std(sample_means)
    sampling_stderr = sampling_std / np.sqrt(sample_size)
    print(f'Sampling Mean: {sampling_mean}')
    print(f'Sampling Variance: {sampling_variance}')
    print(f'Sampling Standard Deviation: {sampling_std}')
    print(f'Sampling Standard Error: {sampling_stderr}')
```

```python
def normal():
    print("Population parameters of normal distribution are : population_mean is 50,standard deviation is 10,size of 10000")
    population_mean = 50
    population_std = 10
    population_size = 10000

    population = np.random.normal(population_mean, population_std, population_size)
    ax=sns.displot(population,kde=True)
    population_mean = np.mean(population)
    population_variance = np.var(population)
    print("Mean and variance of population are:",population_mean,population_variance)
    
    print("Number of samples to take from the population are 1000")
    num_samples = 1000

    print("Size of each sample is 50")
    sample_size = 50 

    sample_means=[]
    for x in range(num_samples):
        sample=np.random.choice(population,sample_size)
        sample_means.append(np.mean(sample))



    sns.displot(sample_means, kde=True)
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.title('Sample Means (Central Limit Theorem using Normal Distribution)')
    plt.grid(True)
    plt.show()
    sampling_mean = np.mean(sample_means)
    sampling_variance = np.var(sample_means)
    sampling_std = np.std(sample_means)
    sampling_stderr = sampling_std / np.sqrt(sample_size)
    print(f'Sampling Mean: {sampling_mean}')
    print(f'Sampling Variance: {sampling_variance}')
    print(f'Sampling Standard Deviation: {sampling_std}')
    print(f'Sampling Standard Error: {sampling_stderr}')
```

```python
def poission():
    print("Population parameters of Poisson distribution are : lambda is 3.5")
    population_lambda = 3.5  

    population = np.random.poisson(population_lambda, size=10000)
    ax=sns.displot(population,kde=True)
    
    population_mean = np.mean(population)
    population_variance = np.var(population)
    print("Mean and variance of population are:",population_mean,population_variance)

    print("Number of samples to take from the population are 1000")
    num_samples = 5000

    print("Size of each sample is 50")
    sample_size = 50

    sample_means=[]
    for x in range(num_samples):
        
        
        sample=np.random.choice(population,sample_size)
        sample_means.append(np.mean(sample))

    sns.displot(sample_means, kde=True)

    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.title('Sample Means (Central Limit Theorem using Poisson Distribution)')
    plt.grid(True)
    plt.show()
    sampling_mean = np.mean(sample_means)
    sampling_variance = np.var(sample_means)
    sampling_std = np.std(sample_means)
    sampling_stderr = sampling_std / np.sqrt(sample_size)
    print(f'Sampling Mean: {sampling_mean}')
    print(f'Sampling Variance: {sampling_variance}')
    print(f'Sampling Standard Deviation: {sampling_std}')
    print(f'Sampling Standard Error: {sampling_stderr}')
```
